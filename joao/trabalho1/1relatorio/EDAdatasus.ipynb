### Importando as bibliotecas que iremos utilizar

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
# configs
sns.set()
pd.set_option('display.float_format', lambda x: '%.5f' % x)
plt.rcParams['figure.figsize'] = [17, 8]
```

## Importando dados dos obitos por desnutrição

```python
des = pd.read_csv('data.csv',';')
des = des.replace('-',0)
# des.index = des['Ano do Óbito']
# des = des.drop('Ano do Óbito',axis=1
des.columns = ['index', '3205 SUL', '3207 METROPOLITANA', '3209 CENTRAL NORTE','Total']
des = des.iloc[:-1,:]
des = des.apply(pd.to_numeric)
```

```python
_ = plt.pie(des.sum()[1:4],labels=['3205 SUL', '3207 METROPOLITANA', '3209 CENTRAL NORTE'],autopct='%.0f%%')
_ = plt.title('Distribuição dos óbitos por macroregião',fontsize=25)
```

## Série historica de desnutrição
Como esperado vamos ver um enorme declinio. Pois a desnutrição foi algo focado globalmente como esforço coletivo.

```python
palette = plt.get_cmap('Set1')
num=0
df = des
for column in df.drop('index', axis=1):
    num+=1
    plt.plot(df['index'], df[column], marker='o', color=palette(num), linewidth=1, alpha=0.9, label=column)
plt.legend(loc=2, ncol=2)
plt.title("Óbitos por desnutrição", loc='left', fontsize=20, fontweight=0, color='Black')
plt.xlabel("Ano")
plt.ylabel("População")
plt.yscale('linear')
plt.ticklabel_format(style='plain', axis='y', scilimits=(0,0))
#Show the graph
plt.show()
```

## Importando dados populacionais

```python
pop = pd.read_csv('datas.csv',';')
colnames = pop.iloc[:,0]
pop2 = pop.transpose().iloc[5:]
pop2.columns = colnames
pop2 = pop2.reset_index()
pop2 = pop2[pop2.iloc[:,0] != '2018']
pop2 = pop2[pop2.iloc[:,0] != '2019']
pop2 = pop2.apply(pd.to_numeric)
```

```python
_ = plt.pie(pop2.iloc[20,:][1:4],labels=['3205 SUL', '3207 METROPOLITANA', '3209 CENTRAL NORTE'],autopct='%.0f%%')
_ = plt.title('Distribuição da população por macroregião',fontsize=25)
```

## Serie historica do crescimento populacional do Espirito Santo

```python
palette = plt.get_cmap('Set1')
num=0
df = pop2
for column in df.drop('index', axis=1):
    num+=1
    plt.plot(df['index'], df[column], marker='o', color=palette(num), linewidth=1, alpha=0.9, label=column)
plt.legend(loc=2, ncol=2)
plt.title("População do espirito santo", loc='left', fontsize=25, fontweight=0, color='Black')
plt.xlabel("Ano")
plt.ylabel("População")
plt.yscale('linear')
plt.ticklabel_format(style='plain', axis='y', scilimits=(0,0))
# Show the graph
plt.show()
```

## % de obitos pela população total

```python
perc = des[des['index']<2018]
perc = perc[perc['index']>1996]
percpop2 = pop2[pop2['index']<2018]

matrix_perc = perc.iloc[:,1:].to_numpy()/percpop2.iloc[:,1:].to_numpy()
x = pd.DataFrame(np.arange(1997,2018))
x["3205 SUL"] = matrix_perc[:,0]
x["3207 METROPOLITANA"] = matrix_perc[:,1]
x["3209 CENTRAL NORTE"] = matrix_perc[:,2]
x.columns = ['index', '3205 SUL', '3207 METROPOLITANA', '3209 CENTRAL NORTE']
```

```python
palette = plt.get_cmap('Set1')
num=0
df = x
for column in df.drop('index', axis=1):
    num+=1
    plt.plot(df['index'], df[column], marker='o', color=palette(num), linewidth=1, alpha=0.9, label=column)
plt.legend(loc=2, ncol=2)
plt.title("Mortes por desnutrição", loc='left', fontsize=12, fontweight=0, color='Black')
plt.xlabel("Ano")
plt.ylabel("População")
plt.yscale('linear')
plt.ticklabel_format(style='plain', axis='y', scilimits=(0,0))
#Show the graph
plt.show()
```
